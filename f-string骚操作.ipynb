```python
# 1. 数字千位分隔符
price = 1234567
print(f"{price:,}")
# 输出：1,234,567
```

```python
# 2. 保留小数+百分比
rate = 0.3789
print(f"{rate:.2%}")
# 输出：37.89%
```

```python
# 3. 对齐补充
name = "Python"
print(f"{name:*<10}")
# 输出：Python****
```

```python
print(f"{name:*>10}")
# 输出：****Python
```

```python
# 4. 日期格式化
from datetime import date
today = date.today()
print(f"{today:%Y年%m月%d日}")
# 输出：2026年06月27日
```

```python
# 5. 直接在f-string里写表达式
print(f"{[x*2 for x in range(5)]}")
# 输出：[0, 2, 4, 6, 8]
```
